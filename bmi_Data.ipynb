```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already read the 'insurance.csv' file into 'df'
df = pd.read_csv('insurance.csv')

# Create a figure with multiple subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 10))  # 2 rows and 2 columns of subplots

# First subplot: Age vs. BMI
axs[0, 0].scatter(df['age'], df['bmi'], alpha=0.5)
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('BMI')
axs[0, 0].set_title('Age vs. BMI')

# Second subplot: Smokers vs. BMI
smokers = df[df['smoker'] == 'yes']
non_smokers = df[df['smoker'] == 'no']
axs[0, 1].scatter(smokers['bmi'], smokers['smoker'], label='Smokers', alpha=0.5, color='red')
axs[0, 1].scatter(non_smokers['bmi'], non_smokers['smoker'], label='Non-Smokers', alpha=0.5, color='blue')
axs[0, 1].set_xlabel('BMI')
axs[0, 1].set_ylabel('Smoker Status')
axs[0, 1].set_title('Smokers vs. BMI')
axs[0, 1].legend()

# Third subplot: BMI vs. Charges
axs[1, 0].scatter(df['bmi'], df['charges'], alpha=0.5)
axs[1, 0].set_xlabel('BMI')
axs[1, 0].set_ylabel('Charges')
axs[1, 0].set_title('BMI vs. Charges')

# Fourth subplot: Children vs. BMI
axs[1, 1].scatter(df['children'], df['bmi'], alpha=0.5)
axs[1, 1].set_xlabel('Children')
axs[1, 1].set_ylabel('BMI')
axs[1, 1].set_title('Children vs. BMI')

# Fifth subplot: Sex vs. BMI
male = df[df['sex'] == 'male']
female = df[df['sex'] == 'female']
axs[2, 0].scatter(male['bmi'], male['sex'], label='Male', alpha=0.5, color='green')
axs[2, 0].scatter(female['bmi'], female['sex'], label='Female', alpha=0.5, color='purple')
axs[2, 0].set_xlabel('BMI')
axs[2, 0].set_ylabel('Sex')
axs[2, 0].set_title('Sex vs. BMI')
axs[2, 0].legend()

# Sixth subplot: Region vs. BMI
northeast = df[df['region'] == 'northeast']
northwest = df[df['region'] == 'northwest']
southeast = df[df['region'] == 'southeast']
southwest = df[df['region'] == 'southwest']
axs[2, 1].scatter(northeast['bmi'], northeast['region'], label='Northeast', alpha=0.5, color='orange')
axs[2, 1].scatter(northwest['bmi'], northwest['region'], label='Northwest', alpha=0.5, color='blue')
axs[2, 1].scatter(southeast['bmi'], southeast['region'], label='Southeast', alpha=0.5, color='red')
axs[2, 1].scatter(southwest['bmi'], southwest['region'], label='Southwest', alpha=0.5, color='green')
axs[2, 1].set_xlabel('BMI')
axs[2, 1].set_ylabel('Region')
axs[2, 1].set_title('Region vs. BMI')
axs[2, 1].legend()


# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.4, hspace=0.5)

#Save the graphs to a PDF
plt.savefig('bmi_plots.pdf')

# Show the plots
plt.show()




    
```
